```python
primekg[['relation','x_index','x_type','x_name','y_index','y_type','y_name']].drop_duplicates(subset='relation').reset_index(drop=True)
```

```python
data = PrimeKG(path = './data')
drug_feature = data.get_features(feature_type = 'drug')
disease_featrue = data.get_features(feature_type = 'disease')
data.to_nx()
data.get_node_list(node_type = 'disease')
```

```python
disease_featrue.query('node_index==28158')[['node_index','mondo_id','mondo_name','group_name_bert','mondo_definition']].drop_duplicates(subset='mondo_name').reset_index(drop=True)
```

```python
subgraph  = primekg.query('x_type=="gene/protein" | y_type=="gene/protein"').copy()
subgraph = subgraph[~subgraph.relation.isin(['anatomy_protein_present','anatomy_protein_absent','cellcomp_protein','molfunc_protein'])]
subgraph = subgraph.drop_duplicates()
subgraph = subgraph.dropna()
subgraph.relation.value_counts()
```

```python
disease_graph = primekg.query('x_type=="disease" and y_type=="disease"')
# y_index==28158
disease_graph = disease_graph.query('x_name=="inflammatory bowel disease" | y_name =="inflammatory bowel disease"' )

subgraph = pd.concat([subgraph, disease_graph])
```

```python
kg = subgraph.loc[:,['x_name','relation','y_name']].copy()
kg.relation.value_counts()
```

```python
kg.to_csv('./output/subgraph.csv',sep='\t')
```

```python
kg_gene = kg.query('relation=="protein_protein"').copy()
kg_gene.relation = kg_gene.relation.str.replace('protein_protein','association')
kg_gene.to_csv('./output/subgraph-gene.csv',sep='\t')
```
