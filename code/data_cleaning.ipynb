```python
import pandas as pd
import os

# read the data
# primekg = pd.read_csv('./data/kg.csv', low_memory=False)
df = pd.read_csv('./data/anti-p40/GSE112366_data.csv')
df_meta = pd.read_csv('./data/anti-p40/GSE112366_metadata.csv')
df.loc[:,'GENE_SYMBOL'] = df.loc[:,'GENE_SYMBOL'].str.split('///')
df = df.explode('GENE_SYMBOL').reset_index(drop=True)
df.loc[:,'GENE_SYMBOL'] = df.loc[:,'GENE_SYMBOL'].apply(lambda x:x.strip())

# only select visit == baseline and treamt == ustekinumab, or visit == control group
df_sample = df_meta[((df_meta['visit'] == 'baseline') & (df_meta['treatment'] == 'ustekinumab')) | (df_meta['visit'] == 'control')]
gsm_list = df_sample.geo_accession.tolist() # target sample list
df_sample.loc[:,['visit','treatment','response']].value_counts(dropna=False)
# step-1. ->design.csv
design = df_sample.loc[:,['geo_accession','visit']].copy()
design.to_csv('./output/design_1.csv',sep='\t',index=False)

# step-2. ->metadata.csv
samples = ['ID','GENE_SYMBOL']

for i in df.columns.tolist()[1:]:
    if i in gsm_list:
        samples.append(i)

clep = df.loc[:,df.columns.isin(samples)].copy()

clep.rename(columns = {'ID':'probe_set_id'},inplace=True)
clep.loc[:,'variance'] = clep.drop(['probe_set_id','GENE_SYMBOL'],axis=1).var(axis=1)

metadata = pd.pivot_table(data = clep, index='GENE_SYMBOL',aggfunc='max').reset_index()

metadata.drop(columns=['variance','probe_set_id'],inplace=True)
metadata.shape
metadata.to_csv('./output/metadata_1.csv',sep='\t',index = False)
```
